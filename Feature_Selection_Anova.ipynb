### regular test train split

```python
y = df.iloc[:,13]
x = df.iloc[:,0:13]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Selecting Features

```python
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(f_classif, k=4)
selector.fit(x_train, y_train)
```

### Top Features

```python
cols = selector.get_support(indices=True)
```

```python
cols
```

### picking subset of train and test

```python
x_train_s = x_train.iloc[:,cols]
x_test_s = x_test.iloc[:,cols]
```

#training with all features using DT

```python
from sklearn.tree import DecisionTreeClassifier
#Initalize the classifier
clf = DecisionTreeClassifier()

clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

### training with selected features

```python
clf.fit(x_train_s, y_train)
y_pred_s = clf.predict(x_test_s)
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred_s))
```
